```python
sales_df['total_revenue']=sales_df['price']*sales_df['quantity']
rev_by_cat=sales_df.groupby('category')['total_revenue'].sum().sort_values(ascending=False).reset_index()

rev_by_cat
```

```python
plt.figure(figsize=(10,6))
plt.bar(rev_by_cat['category'],rev_by_cat['total_revenue'])
plt.title('Total revenue by category')
plt.xlabel('Category')
plt.ylabel('total rev ($)')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
sales_df['revenue']=sales_df['price']*sales_df['quantity']
sales_customers=pd.merge(sales_df,customers_df,on='customer_id', how='inner')
mem_analysis=sales_customers.groupby('membership_level').agg(
    customer_count=('customer_id','nunique'),
    order_count=('order_id','count'),
    total_spent=('revenue','sum'),
    avg_order_val=('revenue','mean')
).sort_values('total_spent', ascending=False)
mem_analysis['avg_soent_per_customer']=mem_analysis['total_spent']/mem_analysis['customer_count']
mem_analysis
```

```python
plt.figure(figsize=(10,6))
plt.bar(mem_analysis.index,mem_analysis['total_spent'])
plt.xlabel('Membership Level')
plt.ylabel('Total Spent ($)')
plt.title('Total spending by membership level')
plt.grid(axis='y')

for     i, v in enumerate(mem_analysis['total_spent']):
    plt.text(i,v+3, f'${v:.2f}', ha='center')
plt.tight_layout()
plt.show()
```

```python
top_customenrs=sales_customers.groupby(['customer_id','name','membership_level']).agg(
    order_count=('order_id','count'),
    total_spent=('revenue', 'sum')
).reset_index().sort_values('total_spent',ascending=False)
top_customenrs
```

```python
gold_member_purchases = sales_customers[sales_customers['membership_level'] == 'Gold']
gold_member_categories = gold_member_purchases.groupby('category').agg(
    order_count=('order_id', 'count'),
    total_quantity=('quantity', 'sum'),
    total_spent=('revenue', 'sum')
).reset_index().sort_values('total_spent', ascending=False)

print("Categories popular with Gold members:")
gold_member_categories
```
